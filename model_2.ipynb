```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.lines as mlines
```

# Read the CSV and Perform Basic Data Cleaning

```python
# Read CSV
df = pd.read_csv("exoplanet_data.csv")

# Drop the null columns where all values are null
df = df.dropna(axis='columns', how='all')

# Drop the null rows
df = df.dropna()
df.head()
```

# Select your features (columns)

```python
# Set features. This will also be used as your x values.
selected_features = df[['koi_disposition', 'koi_period', 'koi_time0bk', 'koi_duration',
                       'koi_depth', 'koi_prad', 'koi_steff', 'koi_slogg', 'koi_srad']]

# Note: Do not select the error margins

selected_features = selected_features.rename(columns = {
    'koi_disposition': "Planetary Status", 
    'koi_period': "Orbit period (d)", 
    'koi_time0bk': "Transit epoch", 
    'koi_duration': "Transit duration (h)",
    'koi_depth': "Transit depth", 
    "koi_prad": "Planet radius",
    'koi_steff': "Stellar temp (K)", 
    'koi_slogg': "Stellar gravity", 
    'koi_srad': "Stellar radius"
})

selected_features
```

```python
# Remove rows where "Planetary Status" is Candidate (unconfirmed status)
selected_features = selected_features.loc[selected_features["Planetary Status"] != "CANDIDATE"]
```

```python
# Get count of Confirmed planets and False Positives
confirmed = selected_features.groupby("Planetary Status")
confirmed["Orbit period (d)"].count()
```

```python
# Convert "Planetary Status" column to integer
selected_features = selected_features.replace({"CONFIRMED": int(1),
                                               "FALSE POSITIVE": int(-1)})

selected_features["Planetary Status"] = selected_features["Planetary Status"].astype("int64")
```

```python
# Inspect planetary radius
print("max radius = ", selected_features["Planet radius"].max())
print("min radius = ", selected_features["Planet radius"].min())
print("median radius = ", selected_features["Planet radius"].median())
```

```python
# Drop rows with extremely large planetary radii (improves scatter plot visualization and also model training)
selected_features = selected_features.loc[selected_features["Planet radius"] <=10000]
```

# Create a Train Test Split

Use `koi_disposition` for the y values

```python
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import MinMaxScaler, OneHotEncoder, LabelBinarizer
from sklearn.model_selection import train_test_split
from sklearn.compose import ColumnTransformer
from sklearn.compose import make_column_transformer
from sklearn.compose import make_column_selector
```

```python
# Set y values (target). 
target = selected_features["Planetary Status"]
target_names = ["CONFIRMED", "FALSE POSITIVE"]
```

```python
X = selected_features[["Orbit period (d)", "Transit epoch", "Transit duration (h)",
                            "Planet radius", "Stellar temp (K)", "Stellar gravity", "Stellar radius"]]

feature_names = X.columns
```

```python
# split the data
X_train, X_test, y_train, y_test = train_test_split(X, target, random_state=42)
```

# Pre-processing

Scale the data using the MinMaxScaler and perform some feature selection

```python
# Reshape the y values to pass through the MinMaxScaler() function
y_train.shape
y_train = y_train.values.reshape(-1, 1)
y_test = y_test.values.reshape(-1, 1)
print(y_train.shape)
print(y_test.shape)
```

```python
# Scale your data
X_minmax = MinMaxScaler(feature_range=(0, 10)).fit(X_train)
y_minmax = MinMaxScaler(feature_range=(0, 10)).fit(y_train)

X_train_scaled = X_minmax.transform(X_train)
X_test_scaled = X_minmax.transform(X_test)
y_train_scaled = y_minmax.transform(y_train)
y_test_scaled = y_minmax.transform(y_test)
```

```python
# Plot and show scatter plot data
fig1 = plt.figure(figsize=(12, 11))
axes1 = fig1.add_subplot(2, 2, 1)
axes2 = fig1.add_subplot(2, 2, 2)
axes3 = fig1.add_subplot(2, 2, 3)
axes4 = fig1.add_subplot(2, 2, 4)

def set_plot(axis, x_label, y_label, x_data, y_data):
    axis.set_title(f"{x_label} vs. {y_label}")
    axis.set_xlabel(f"{x_label}")
    axis.set_ylabel(f"{y_label}")
    plt.xlim(0, 10)
    plt.ylim(0, 10)
    return axis.scatter(X_train_scaled[:, x_data], X_train_scaled[:, y_data], cmap="bwr")

orbit_period = 0
transit_epoch = 1
transit_duration = 2
planet_radius = 3
stellar_gravity = 5
stellar_radius = 6

set_plot(axes1, "Orbit Period", "Planet radius", orbit_period, planet_radius);
set_plot(axes2, "Transit Epoch", "Transit Duration", transit_epoch, transit_duration);
set_plot(axes3, "Stellar Gravity", "Orbit Period", stellar_gravity, orbit_period);
set_plot(axes4, "Orbit Period", "Transit Duration", orbit_period, transit_duration);

plt.show()
```

# Train the Model



```python
# Create a kmeans model using k = 2
from sklearn.cluster import KMeans
model1 = KMeans(n_clusters=2)
model1.fit(X_train_scaled)

# Use the data to predict the clusters
# save the predictions as `predicted_clusters`
predicted_clusters_trained = model1.predict(X_train_scaled)
```

```python
# Print the cluster centers and cluster labels
centers_trained = model1.cluster_centers_
labels_trained = model1.labels_
print(centers_trained)
print(labels_trained)
```

```python
# Define function to plot data.

def show_cluster_plots(axis, x_label, y_label, x_data, y_data, data, centers, clusters):
    axis.set_title(f"{x_label} vs. {y_label}")
    axis.set_xlabel(f"{x_label}")
    axis.set_ylabel(f"{y_label}")
    plt.xlim(0, 10)
    plt.ylim(0, 10)    
    axis.scatter(data[:, x_data], data[:, y_data], c=clusters, s=50, cmap="viridis")
    axis.scatter(centers[:, x_data], centers[:, y_data], c='red', s=200, alpha=0.5)
    return axis
```

Plot the models for train data

```python
# Plot and show scatter plot data
fig2 = plt.figure(figsize=(12, 11))
axes5 = fig2.add_subplot(2, 2, 1)
axes6 = fig2.add_subplot(2, 2, 2)
axes7 = fig2.add_subplot(2, 2, 3)
axes8 = fig2.add_subplot(2, 2, 4)

show_cluster_plots(axes5, "Orbit Period", "Planet radius", orbit_period, planet_radius, X_train_scaled, centers_trained, predicted_clusters_trained);
show_cluster_plots(axes6, "Transit Epoch", "Transit Duration", transit_epoch, transit_duration, X_train_scaled, centers_trained, predicted_clusters_trained);
show_cluster_plots(axes7, "Orbit Period", "Stellar Gravity", orbit_period, stellar_gravity, X_train_scaled, centers_trained, predicted_clusters_trained);
show_cluster_plots(axes8, "Orbit Period", "Transit Duration", orbit_period, transit_duration, X_train_scaled, centers_trained, predicted_clusters_trained);

fig2.savefig("Images/Trained_Data_Clusters.png")

plt.show()
```

```python
# Run on test data
model2 = KMeans(n_clusters=2)
model2.fit(X_test_scaled)
predicted_clusters_test = model2.predict(X_test_scaled)
```

```python
# Print the cluster centers and cluster labels
centers_test = model2.cluster_centers_
labels_test = model2.labels_
print(centers_test)
print(labels_test)
```

Plot the model for test data

```python
# Plot the clusters
# Plot and show scatter plot data
fig3 = plt.figure(figsize=(12, 11))
axes9 = fig3.add_subplot(2, 2, 1)
axes10 = fig3.add_subplot(2, 2, 2)
axes11 = fig3.add_subplot(2, 2, 3)
axes12 = fig3.add_subplot(2, 2, 4)

show_cluster_plots(axes9, "Orbit Period", "Planet radius", orbit_period, planet_radius, X_test_scaled, centers_test, predicted_clusters_test);
show_cluster_plots(axes10, "Transit Epoch", "Transit Duration", transit_epoch, transit_duration, X_test_scaled, centers_test, predicted_clusters_test);
show_cluster_plots(axes11, "Orbit Period", "Stellar Gravity", orbit_period, stellar_gravity, X_test_scaled, centers_test, predicted_clusters_test);
show_cluster_plots(axes12, "Orbit Period", "Transit Duration", orbit_period, transit_duration, X_test_scaled, centers_test, predicted_clusters_test);

fig3.savefig("Images/Test_Data_Clusters.png")
```

```python
print(f"Training Data Score: {model1.score(X_train_scaled, y_train_scaled)}")
print(f"Testing Data Score: {model1.score(X_test_scaled, y_test_scaled)}")

print(f"Training Data Score: {model2.score(X_train_scaled, y_train_scaled)}")
print(f"Testing Data Score: {model2.score(X_test_scaled, y_test_scaled)}")
```

# Hyperparameter Tuning

Use `GridSearchCV` to tune the model's parameters

```python
# Create the GridSearchCV model
from sklearn.pipeline import make_pipeline
from sklearn.svm import SVC 
from sklearn import set_config
set_config(display="diagram")

pipe = make_pipeline(model1)
pipe
```

```python
# Create the GridSearch estimator along with a parameter object containing the values to adjust
from sklearn.model_selection import GridSearchCV
param_grid = {'kmeans__C': [1, 5, 10, 50]}
grid = GridSearchCV(pipe, param_grid, verbose=3)
```

```python
# Train the model with GridSearch
# Fit the model using the grid search estimator. 
# Note: This operation takes about 30-35 minutes to run.
grid.fit(X_train_scaled)
```

```python
# Print the best params
print(grid.best_params_)
print(grid.best_score_)
```

```python
# Make predictions with the hypertuned model
predictions = grid.predict(X_test_scaled)

# Calculate classification report
from sklearn.metrics import classification_report
print(classification_report(y_test_scaled, predictions, target_names=target_names))
```
